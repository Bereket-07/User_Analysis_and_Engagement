#### geting the 10 handeset used by the customers

```python
top_ten_handset_type = df1['Handset Type'].value_counts().sort_values(ascending=False).head(10)
top_ten_handset_type
```

```python
top_ten_handset_type.plot(kind='bar' , color='skyblue')
plt.title("Top ten handset chose by the user")
plt.xlabel('types of handsets(Mobile phones)')
plt.ylabel('the number of handsets used by our user')
plt.xticks(rotation=45, ha='right')
plt.show()
```

#### Top 3 manufacturers

```python
top3_manufacturers = df1['Handset Manufacturer'].value_counts().sort_values(ascending=False).head(3)
top3_manufacturers
```

```python
top3_manufacturers.plot(kind='bar' , color='lightgreen')
plt.xlabel('manufacturer name')
plt.ylabel('number of handset manufactured by the company')
plt.title("Top three manufacrurers")
plt.show()
```
