Distribution of Average Throughput per Handset Type

In the first graph, we see the distribution of average throughput per handset type. The histogram shows that the majority of the throughput values are concentrated at the lower end, with a significant number of data points around the 0 to 1000 kbps range. As the throughput increases, the frequency of data points decreases, with a long tail extending towards higher throughput values.
Interpretation:

    Skewed Distribution: The distribution is heavily right-skewed, indicating that most handset types experience lower average throughput, while only a few achieve high throughput values.
    Possible Bottlenecks: The concentration of low throughput values suggests that there may be network or device-related bottlenecks affecting the performance of most handsets.
    Outliers: The presence of a long tail could indicate the existence of outliers or special cases where certain handsets achieve significantly higher throughput, possibly due to better hardware, network conditions, or data usage patterns.

Distribution of Average TCP Retransmission per Handset Type

In the second graph, the distribution of average TCP retransmission per handset type is displayed. The histogram has a bimodal distribution with two prominent peaks. One peak occurs near zero, and another around a higher retransmission value, indicating two distinct groups in the dataset.
Interpretation:

    Bimodal Distribution: The bimodal nature suggests that there are two separate groups or behaviors in the dataset. One group of handset types experiences almost no retransmissions, while another group shows significantly higher retransmissions.
    Network Reliability Issues: The high retransmission values in one group could indicate issues with network reliability or handset performance, leading to frequent retransmissions.
    Zero Retransmissions: The high frequency of zero or near-zero retransmissions might indicate that for many handsets, the network conditions are optimal, resulting in minimal data packet loss.

Overall, the visualizations suggest that the majority of handset types experience lower average throughput and are generally reliable with minimal retransmissions. However, there are outliers or special cases that warrant further investigation to improve network performance and user experience.
